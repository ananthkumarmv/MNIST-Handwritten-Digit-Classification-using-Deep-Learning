```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import cv2
from PIL import Image
import tensorflow as tf
from tensorflow import keras
from keras.datasets import mnist
from tensorflow.math import confusion_matrix
```

```python
tf.random.set_seed(3)
```

```python
# Loading MNIST dataset from keras.datasets
(x_train, y_train), (x_test, y_test) = mnist.load_data()
```

```python
type(x_train)
```

```python
print( x_train.shape)
print(x_test.shape) 
print(y_train.shape)
print(y_test.shape)
```

Image Labels

```python
# unique values in y_train
print(np.unique(y_train))

# unique values in y_test
print(np.unique(y_test))
```

```python
# Scaling the values
x_train = x_train/255
x_test = x_test/255
```

```python
# displaying image after scaling
plt.imshow(x_train[50])
plt.show()

# corresponding label
print(y_train[50])
```

## Building the Neural Network

```python
# Setting up the layers of the Neural Network
model = keras.Sequential([
    keras.layers.Flatten(input_shape=(28, 28)),
    keras.layers.Dense(50, activation = 'relu'),
    keras.layers.Dense(50, activation = 'relu'),
    keras.layers.Dense(10, activation='softmax')
])
```

```python
# compiling the


model.compile(optimizer='adam', 
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
# Training the Neural Network

model.fit(x_train, y_train, epochs=10)
```

```python
#Evaluating the trained model
model.evaluate(x_test, y_test)
```

```python
y_pred = model.predict(x_test)
```

```python
# model.predict() gives probability of each class for that data point
y_pred[10]
```

```python
y_pred.shape
```

```python
# Converting prediction probabilities to class label for all test data points

y_pred_labels = [np.argmax(i) for i in y_pred]
```

```python
plt.imshow(x_test[20])
plt.show()

print(y_pred_labels[20])
```

Confusion matrix

```python
conf_mat = confusion_matrix(y_test, y_pred_labels)
```

```python
print(conf_mat)
```

```python
plt.figure(figsize=(15,7))
sns.heatmap(conf_mat, annot=True, fmt='d', cmap='Blues')
plt.xlabel('Predicted Labels')
plt.ylabel('True Labels')
```

# Predictive system

```python
#reading image
input_image_path = 'Images/Seven.webp'
input_image = cv2.imread(input_image_path)

# Converting to grayscale
grayscale = cv2.cvtColor(input_image, cv2.COLOR_RGB2GRAY)

# Resizing image
input_image_resize = cv2.resize(grayscale, (28, 28))

# Normalizing
input_image_resize = input_image_resize/255

# Reshaping
image_reshaped = np.reshape(input_image_resize, [1, 28, 28])

# Predicting the probability
input_prediction = model.predict(image_reshaped)

input_pred_label = np.argmax(input_prediction)

print(input_pred_label)
```
